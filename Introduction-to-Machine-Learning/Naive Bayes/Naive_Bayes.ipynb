üìå The main aim of classification is, to find the class corresponding to a given feature set. The **Naive Bayes** classifier algorithm is based on a famous theorem called **‚ÄúBayes theorem‚Äù** which is centered on conditional probability. Conditional probability is the probability of an event ‚ÄòA‚Äô happening given that another event ‚ÄòB‚Äô has already happened. For example, consider event ‚ÄòA‚Äô to be ‚Äúhaving a fever‚Äù, and event ‚ÄòB‚Äô to be ‚Äúinfected with Covid-19‚Äù. With conditional probability, we can ask the question: what is the chance of having a fever given that you have been infected with Covid-19.

üìå The Bayes‚Äô theorem is an extension of conditional probability. In a sense, it allows us to use reverse reasoning. The Naive Bayes algorithm does the same for the class and its features. Instead of calculating the probability of a feature belonging to a class, it approaches the issue from another angle. 

üìå There are three types of Naive Bayes Classifiers in sklearn; **Bernoulli Naive Bayes**, **Multinomial Naive Bayes** and **Gaussian Naive Bayes**. We use Bernoulli Naive Bayes when our data is binary like true or false, yes or no and so on. We use Multinomial Naive Bayes when we have discrete values such as number of family members or, pages in a book. We use Gaussian Naive Bayes when all of our features are continuous variables, like temperature or height. Let‚Äôs take the dataset on tumors from our classification session, which only has continuous variables. We use the Gaussian Naive Bayes algorithm. As always, we start with importing the Pandas library for reading our data file. Then we can read and display the first few rows of the dataset.

```python
import pandas as pd
```

```python
dataset = pd.read_csv("breast-cancer.csv")
dataset.head()
```

üìå we need to convert our target variable from categorical to numerical type using label encoding. Then, we can continue with defining our features and a target.kalƒ±n metin

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
dataset["diagnosis"] = labelencoder.fit_transform(dataset["diagnosis"].values) 
```

```python
X = dataset.drop("diagnosis", axis =1)
y = dataset["diagnosis"]
```

üìå After we define the features and a target, we can split them into train and test.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

üìå Now, the most exciting part! Let‚Äôs create our model, Gaussian Naive Bayes, teach some hidden patterns to it with training data, and finally use it to make predictions.

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, y_train)
```

üìå Now, we can check the strength of our predictions.

```python
predictions = model.predict(X_test)
predictions
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, predictions)
```

üìå Looking at the results, we have accuracy and precision above 90% which can be considered as a good result.

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predictions))
```
